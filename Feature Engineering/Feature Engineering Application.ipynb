```python
def woe_iv(data : pd.DataFrame, variable_name : str, br : str = 'Bin_Range') -> tuple[float, pd.DataFrame]:
    # data = input dataframe that has already been binned for the predictor variable_name
    # variable_name = predictor variable, not used in this function
    working_data = data.copy()

    # missing values have been assigned NaN when binning with pd.qcut
    # rename this bin as 'Missing' to consider into WOE calculation
    working_data[br] = working_data[br].astype('category')
    miss = 'Missing'
    if miss not in working_data[br].cat.categories:
        working_data[br] = working_data[br].cat.add_categories(miss)
    working_data[br] = working_data[br].fillna(miss)

    variable_data = pd.DataFrame()
    variable_data[br] = working_data[br].cat.categories
    # everything is faster with numpy
    gb = working_data.groupby(br, observed = False)['TARGET']
    c = gb.count().to_numpy()
    variable_data['Count'] = c
    se = gb.sum().to_numpy()
    variable_data['Events'] = se
    ne =  c - se
    variable_data['Non_Events'] = ne
    pe = se / se.sum()
    variable_data['%_of_Events'] = pe
    pne = ne / ne.sum()
    variable_data['%_of_Non_Events'] = pne
    # suppress warning if taking log of zero
    with np.errstate(divide = 'ignore', invalid = 'ignore'):
        woe = np.log(pne) - np.log(pe)
    variable_data['WOE'] = woe
    variable_data['IV'] = (pne - pe) * woe

    IV = variable_data['IV'].sum(skipna = True)
    return IV, variable_data
```

```python
def woe_iv_plot(data : pd.DataFrame, variable_name : str, br : str = 'Bin_Range'):
    # data = input dataframe that has already been binned for the predictor variable_name
    # variable_name = name of predictor variable that has been binned
    IV, variable_data = woe_iv(data, variable_name, br)
    print(f'IV for {variable_name} with {variable_data.shape[0]} bins = {IV:.4f}')
    fig, ax1 = plt.subplots(figsize = (18, 6))
    woe_col = variable_data['WOE']
    sns.lineplot(data = woe_col.iloc[:-1], marker = 'o', sort = False, ax = ax1)
    plt.xticks(rotation = 45, horizontalalignment = 'right')
    # create the scatter plot to plot the Missing WOE data point, if there is any
    last_point = woe_col.shape[0] - 1
    y = woe_col[last_point]
    if not pd.isna(y):
        sns.scatterplot(x = [last_point], y = [y], color = 'b', s = 30, ax = ax1)
    # plot the bin counts
    ax2 = ax1.twinx()
    sns.barplot(x = variable_data[br] , y = variable_data['Count'], alpha = .5, ax = ax2, color = 'deepskyblue')
    plt.title(f'WOE plot for {variable}')
    plt.show()
```

## Feature Engineering
### Credit-to-Income Ratio
The Credit-to-Income Ratio feature represents the ratio of the credit amount of the loan to the total income of the client. It provides insight into the borrower's financial leverage and ability to manage debt relative to their income.

### LTV (Loan-to-Value) Ratio
The LTV (Loan-to-Value) Ratio feature measures the ratio of the credit amount of the loan to the value of the goods for which the loan is given. It indicates the percentage of the property value that is being financed by the loan. A higher LTV ratio indicates higher risk for the lender.

### Document Provided Count
The Document Provided Count feature counts the total number of documents provided by the client during the loan application process. It serves as an indicator of the completeness of the application documentation.

### Social Surroundings Default
The Social Surroundings Default feature aggregates information on the number of observations of the client's social surroundings with observable default in 30 and 60 days past due (DPD). It provides insight into the client's social environment and potential credit risk.

### Property Ownership Status
The Property Ownership Status feature categorizes clients based on their ownership status of car and realty. It distinguishes between clients who own both, only a car, only realty, or neither, providing insight into their asset ownership profile.

```python
# Credit-to-Income Ratio
df['Credit_to_Income_Ratio'] = df['AMT_CREDIT'] / df['AMT_INCOME_TOTAL']

# LTV (Loan-to-Value) Ratio
df['LTV'] = df['AMT_CREDIT'] / df['AMT_GOODS_PRICE']
df['LTV'].fillna(0, inplace=True)  # Replace NaN values with 0

# Document Provided Count
document_flags = [col for col in df.columns if col.startswith('FLAG_DOCUMENT_')]
df['Document_Provided_Count'] = df[document_flags].sum(axis=1)

# Social Surroundings Default
df['Social_Surroundings_Default'] = df['OBS_30_CNT_SOCIAL_CIRCLE'] + df['OBS_60_CNT_SOCIAL_CIRCLE']

# Property Ownership Status
df['Property_Ownership_Status'] = 'None'
df.loc[(df['FLAG_OWN_CAR'] == 'Y') & (df['FLAG_OWN_REALTY'] == 'Y'), 'Property_Ownership_Status'] = 'Owns Car and Realty'
df.loc[(df['FLAG_OWN_CAR'] == 'Y') & (df['FLAG_OWN_REALTY'] == 'N'), 'Property_Ownership_Status'] = 'Owns Car, No Realty'
df.loc[(df['FLAG_OWN_CAR'] == 'N') & (df['FLAG_OWN_REALTY'] == 'Y'), 'Property_Ownership_Status'] = 'Owns Realty, No Car'
```

```python
def woe_iv_categorical(data: pd.DataFrame, variable_name: str, br: str = 'Bin_Range') -> tuple[float, pd.DataFrame]:
    working_data = data.copy()

    # Ensure the Bin_Range column is categorical
    working_data[br] = working_data[br].astype('category')
    
    # Add 'Missing' category if not present and handle missing values
    miss = 'Missing'
    if miss not in working_data[br].cat.categories:
        working_data[br] = working_data[br].cat.add_categories(miss)
    working_data[br] = working_data[br].fillna(miss)

    # Create a DataFrame to store variable data
    variable_data = pd.DataFrame()
    variable_data[br] = working_data[br].cat.categories

    # Group by Bin_Range and calculate counts of events and non-events
    gb = working_data.groupby(br, observed=False)['TARGET']
    c = gb.count().to_numpy()
    variable_data['Count'] = c
    se = gb.sum().to_numpy()
    variable_data['Events'] = se
    ne = c - se
    variable_data['Non_Events'] = ne

    # Calculate percentage of events and non-events
    pe = se / se.sum()
    pne = ne / ne.sum()
    variable_data['%_of_Events'] = pe
    variable_data['%_of_Non_Events'] = pne

    # Calculate WOE and IV
    with np.errstate(divide='ignore', invalid='ignore'):
        woe = np.log(pne) - np.log(pe)
    variable_data['WOE'] = woe
    variable_data['IV'] = (pne - pe) * woe

    # Sum IV to get total IV
    IV = variable_data['IV'].sum(skipna=True)

    return IV, variable_data
```

```python
variables = ["Document_Provided_Count", "Social_Surroundings_Default", "Property_Ownership_Status"]

br = 'Bin_Range'  # Since we won't be binning categorical variables, you can keep the same name if desired

res = {}
for i in variables:
    df_data_working = new_df.copy()
    df_data_working = df_data_working[[i, 'TARGET']]
    # Since the variable is categorical, no need to bin it
    # You can directly use the categorical variable itself
    df_data_working[br] = df_data_working[i]

    resIV = woe_iv_categorical(df_data_working, i, br)[0].round(4)
    res[i] = resIV

print(res)
```
